```python
import pandas as pd
import numpy as np

# Step 0: Load dataset
df = pd.read_csv("data.csv")
```

```python
#data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#histogram
plt.figure(figsize=(15,10))

#histogram for duration
(plt.subplot(2,2,1))
sns.histplot(df['Duration'],bins=30 , kde=True)
plt.title('Duration Distribution')
```

```python
# Correlation heatmap
corr = df.corr()
plt.figure(figsize=(10, 8))  # Optional: Adjust the figure size
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Exercise Histogram for Calories and Max Pulse
# Drop rows with missing values in relevant columns
df_clean = df.dropna(subset=['Maxpulse', 'Calories'])

# Set up side-by-side plots
plt.figure(figsize=(12, 5))

# Histogram for Maxpulse
plt.subplot(1, 2, 1)
plt.hist(df_clean['Maxpulse'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Maxpulse')
plt.xlabel('Maxpulse')
plt.ylabel('Frequency')
plt.grid(True)

# Histogram for Calories
plt.subplot(1, 2, 2)
plt.hist(df_clean['Calories'], bins=20, color='salmon', edgecolor='black')
plt.title('Histogram of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(True)

# Layout and display
plt.tight_layout()
plt.show()
```

```python

# Drop rows with missing values
df_clean = df.dropna(subset=['Maxpulse', 'Calories'])

# Combined histogram
plt.figure(figsize=(10, 6))
plt.hist(df_clean['Maxpulse'], bins=20, alpha=0.5, label='Maxpulse', color='blue', edgecolor='black')
plt.hist(df_clean['Calories'], bins=20, alpha=0.5, label='Calories', color='orange', edgecolor='black')

plt.title('Combined Histogram of Maxpulse and Calories')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
